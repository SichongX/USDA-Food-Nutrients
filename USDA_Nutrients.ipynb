```python
# If we want to join using the key columns, 
# we need to set key to be the index in both df and other. 
# The joined DataFrame will have key as its index.

USDA = USDA_Macronutrients.set_index("ID").join(USDA_Micronutrients.set_index("ID"),how = "inner").sort_values(by = "ID")
USDA
```

```python
# Remove the records with missing values of 4 or more vectors from the the new data frame

USDA = USDA.dropna(axis = 0, thresh = len(USDA.columns)-4) 
# thresh = # of non-NA values
# inplace
```

```python
USDA
# Col: Sodium, Potassium are mixed with different data type, contian int and string
# Need to convert the data type first
```

```python
# Convert data type
# Series.apply(func, convert_dtype=True, args=(), **kwargs)
USDA.loc[:,"Sodium"] = USDA["Sodium"].apply(lambda x: str(x).replace(",", "")).astype(float)
USDA.loc[:,"Potassium"] = USDA["Potassium"].apply(lambda x: str(x).replace(",", "")).astype(float)
USDA
```

```python
USDA[["Sugar","VitaminE","VitaminD"]] = USDA[["Sugar","VitaminE","VitaminD"]].fillna(USDA.mean())
```

```python
USDA = USDA.dropna()
USDA
```

```python
USDA["HighPo"] = (USDA["Protein"] > USDA["Protein"].mean()).astype(int)
USDA["HighFat"] = (USDA["TotalFat"] > USDA["TotalFat"].mean()).astype(int)
USDA["HighCar"] = (USDA["Carbohydrate"] > USDA["Carbohydrate"].mean()).astype(int)
USDA["HighSo"] = (USDA["Sodium"] > USDA["Sodium"].mean()).astype(int)
USDA["HighCho"] = (USDA["Cholesterol"] > USDA["Cholesterol"].mean()).astype(int)
USDA["HighSugar"] = (USDA["Sugar"] > USDA["Sugar"].mean()).astype(int)
```

```python
# Find how many foods has both high calories and high sugar
USDA.groupby(by = "HighCals").get_group(1).groupby(by = "HighSugar").get_group(1)[["Description"]]
```

```python
# Add a new column "HealthCheck", if the food has 3 or more high records, the food will be labeled with "Failed", otherwise "Pass"
def check(df):
    for i in df.index:
        if sum(df.loc[i,["HighCals","HighSo","HighPo","HighFat","HighCho","HighSugar"]].values) >= 3:
            df.loc[i,"HeathCheck"] = "Failed"
        else:
            df.loc[i,"HeathCheck"] = "Pass"
check(USDA) 
USDA
```

```python
# Alternative method
USDA["High"] = USDA["HighCals"] + USDA["HighSo"] + USDA["HighPo"] + USDA["HighFat"] + USDA["HighCho"] + USDA["HighSugar"]
USDA['HealthCheck1'] = USDA['High'] < 3
USDA
```
